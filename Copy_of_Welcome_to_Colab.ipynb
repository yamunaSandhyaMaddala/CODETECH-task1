```
import pandas as pd

df = pd.read_csv("sales_transactions_regression.csv")
print(df.iloc[:, :4].isnull().sum())
```

```
df['transaction_id'] = df['transaction_id'].fillna(df['transaction_id'].mode()[0])
print(df['transaction_id'])
print(df.iloc[:,0].isnull())
```

```
df['region'] = df['region'].fillna(df['region'].mode()[0])
print(df['region'])
print(df.iloc[:,0].isnull())
```

```
df['product_id'] = df['product_id'].fillna(df['product_id'].mode()[0])
print(df['product_id'])
print(df.iloc[:,0].isnull().sum())
```

```
df['product_id'] = df['product_id'].fillna(df['product_id'].median())
print(df['product_id'])
print(df.iloc[:,0].isnull().sum())
```

```
df['product_id'] = df['product_id'].fillna(df['product_id'].mean())
print(df['product_id'])
print(df.iloc[:,0].isnull().sum())
```

```
df['product_id.1'] = df['product_id.1'].fillna(df['product_id.1'].mode()[0])
print(df['product_id.1'])
print(df.iloc[:,0].isnull().sum())
```

```
df['product_id.1'] = df['product_id.1'].fillna(df['product_id.1'].median())
print(df['product_id.1'])
print(df.iloc[:,0].isnull().sum())
```

```
df['product_id.1'] = df['product_id.1'].fillna(df['product_id.1'].mean())
print(df['product_id.1'])
print(df.iloc[:,0].isnull().sum())
```

```
df['transaction_id'] = df['transaction_id'].fillna(method='ffill')
print(df['transaction_id'].isnull().sum())
```

```
df['transaction_id'] = df['transaction_id'].fillna(method='bfill')
print(df['transaction_id'].isnull().sum())
```

```
df['region'] = df['region'].fillna(method='bfill')
print(df['region'].isnull().sum())
```

```
df['region'] = df['region'].fillna(method= 'ffill')
print(df['region'].isnull().sum())
```

```
df['product_id'] = df['product_id'].fillna(method= 'ffill')
print(df['product_id'].isnull().sum())
```

```
df['product_id'] = df['product_id'].fillna(method= 'bfill')
print(df['product_id'].isnull().sum())
```

```
#It shows how the values of product_id are distributed (their frequency and density) in a clean, visually appealing histogram.
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))

sns.histplot(df['product_id'], bins=20, kde=True, color='skyblue')

plt.title("Distribution of Product ID")
plt.xlabel("Product ID")
plt.ylabel("Frequency")

plt.show()
```

```
#This code creates a **count plot** for the **first column** of a DataFrame (`df`) using Seaborn. It shows how often each unique value appears, adds a title and axis labels, rotates the x-axis labels for readability, and displays the plot.
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x=df.iloc[:, 0])
plt.title(f"Count Plot of {df.columns[0]}")
plt.xlabel(df.columns[0])
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```
#This code plots a count plot for the region column using Seaborn. It displays how many records or transactions belong to each region, helping visualize the frequency distribution of regions in the dataset. The x-axis shows region names, and the y-axis shows their counts, with labels rotated for better readability.
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x=df['region'])
plt.title("Count Plot of Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

The first four columns show each transactions ID, region, sales amount, and discount percentage. From these, we observe that transactions vary across regions, sales amounts differ widely, and discounts influence sales, with some missing values present.
